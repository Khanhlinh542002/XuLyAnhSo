```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v2 as iio
from scipy import ndimage

image = iio.imread('output/starfruit.jpg') 
if image.ndim == 3:
    # Chuyển sang ảnh xám nếu là ảnh màu
    image = np.mean(image, axis=2)
    # Chuyển thành nhị phân
    image = (image > 128).astype(np.uint8)

# Dilation và Erosion với kernel (3, 3)
structure1 = np.ones((3,3))
dilated1 = ndimage.binary_dilation(image, structure=structure1)
eroded1 = ndimage.binary_erosion(image, structure=structure1)

# Dilation và Erosion với kernel (5, 5)
structure2 = np.ones((5,5))
dilated2 = ndimage.binary_dilation(image, structure=structure2)
eroded2 = ndimage.binary_erosion(image, structure=structure2)

# Hiển thị kết quả
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.title('Ảnh gốc')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2,3,2)
plt.title('Dilation (3x3)')
plt.imshow(dilated1, cmap='gray')
plt.axis('off')

plt.subplot(2,3,3)
plt.title('Erosion (3x3)')
plt.imshow(eroded1, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title('Dilation (5x5)')
plt.imshow(dilated2, cmap='gray')
plt.axis('off')

plt.subplot(2,3,6)
plt.title('Erosion (5x5)')
plt.imshow(eroded2, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
